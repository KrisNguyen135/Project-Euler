# Project Euler 686: Powers of Two

Link to the problem prompt: [https://projecteuler.net/problem=686](https://projecteuler.net/problem=686)

Consider a positive integer $j$ such that the first three digits of $2^j$ in base 10 are exactly $123$. Then there exists a positive integer $n$ such that:

$$123 \times 10^n < 2^j < 124 \times 10^n$$

Taking the log in base 10, we have:

$$\frac{n + \log_{10}(123)}{\log_{10}(2)} < j < \frac{n + \log_{10}(124)}{\log_{10}(2)}$$

From here an iterative algorithm can be used to find the numbers $n$ that satisfy the equation for some $j$. Specifically, we need to find values of $n$ such that the rounded value of the LHS is lower than the rounded value of the RHS (which necessitates the existence of at least one integer value of $j$).

To compute $p(123, 678910)$, we grow the list of $j$ values with increasing $n$ until the size of the list reaches $678,910$. This is implemented below by the `find_next_js()` function.

```
from math import sqrt, log10


lower_c = log10(123)
upper_c = log10(124)
divider = log10(2)

js = []
temp_n = 1

def find_next_js(limit):
    global js
    global temp_n

    while len(js) < limit:
        # print(temp_n)
        temp_n += 1

        lower = int((temp_n + lower_c) / divider)
        upper = int((temp_n + upper_c) / divider)

        if lower < upper:
            print(f'*** {temp_n} ***')
            print(f'Current length: {len(js)}')
            print()

            js += [i for i in range(lower + 1, upper + 1)]
```

```
find_next_js(678910)
```
